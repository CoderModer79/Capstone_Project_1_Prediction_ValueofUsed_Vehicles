### Correlation Map

```python
def plot_correlation_map( df ):
    corr = df.corr()
    _ , ax = plt.subplots( figsize =( 12 , 10 ) )
    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )
    _ = sns.heatmap(
        corr, 
        cmap = cmap,
        square=True, 
        cbar_kws={ 'shrink' : .9 }, 
        ax=ax, 
        annot = True, 
        annot_kws = { 'fontsize' : 12 }
    )
plot_correlation_map(dfout)
```

### Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

forest = RandomForestRegressor(n_estimators=100,random_state=0)
forest.fit(X_train,y_train)

print('Accuracy on the training subset: {:.3f}'.format(forest.score(X_train, y_train)))
print('Accuracy on the test subset: {:.3f}'.format(forest.score(X_test, y_test)))
```

```python
from sklearn.model_selection import GridSearchCV

rf = RandomForestRegressor()

param_grid = { "criterion" : ["mse"]
              , "min_samples_leaf" : [3]
              , "min_samples_split" : [3]
              , "max_depth": [10]
              , "n_estimators": [500]}
gs = GridSearchCV(estimator=rf, param_grid=param_grid, cv=2, n_jobs=-1, verbose=1)
gs = gs.fit(X_train, y_train)

print(gs.best_score_)
print(gs.best_params_)
```

## Label Encoding (Preprocessing)

```python
dfout_lbl = dfout.copy()

from sklearn import preprocessing

labels = ['name', 'vehicleType', 
       'gearbox', 'model', 'fuelType', 'brand',
       'notRepairedDamage']

for lbl in labels:
    dfout_lbl[lbl+'_Nmrc'] = preprocessing.LabelEncoder().fit(dfout_lbl[lbl]).transform(dfout_lbl[lbl])

dfout_lbl.drop(labels,axis=1, inplace=True)
#gearboxLbl = preprocessing.LabelEncoder()

#print(gearboxLbl.fit(dfout.gearbox).transform(dfout.gearbox)[i])
```

```python
dfout_lbl['dateCrawled'] = pd.to_numeric(dfout_lbl['dateCrawled'])
dfout_lbl['dateCreated'] = pd.to_numeric(dfout_lbl['dateCreated'])
dfout_lbl['lastSeen'] = pd.to_numeric(dfout_lbl['lastSeen'])



dfout_lbl.columns
```

```python
X_trainNmrc, X_testNmrc, y_trainNmrc, y_testNmrc = train_test_split(dfout_lbl.drop('price',axis=1),dfout_lbl['price'],test_size=.33,random_state=42)
forestNmrc = RandomForestRegressor(n_estimators=100,random_state=0)
forestNmrc.fit(X_trainNmrc,y_trainNmrc)

print('Accuracy on the training subset: {:.3f}'.format(forestNmrc.score(X_trainNmrc, y_trainNmrc)))
print('Accuracy on the test subset: {:.3f}'.format(forestNmrc.score(X_testNmrc, y_testNmrc)))
```
